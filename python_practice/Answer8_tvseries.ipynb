```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import requests
```

**Loading data**

```python
#Retrieve JSON data from URL
response = requests.get("http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes")
data = response.json()
```

```python
# Extract relevant data from the JSON
episode_data = data.get('_embedded').get("episodes")
```

```python
df = pd.DataFrame(episode_data)
```

**Categorical features**

**"airdate"**

```python
# Convert the 'airdate' column to datetime type
df["airdate"] = pd.to_datetime(df["airdate"])
```

**"airtime"**

```python
# Convert the 'Time' column to datetime format
df["airtime"] = pd.to_datetime(df["airtime"], format = '%H:%M')
```

```python
# Convert the 'Time' column to 12-hour format
df['airtime'] = df['airtime'].dt.strftime('%I:%M %p')
```

**"rating"**

```python
df["rating"] = df["rating"].apply(lambda x: x["average"])
```

**"image"**

```python
df["medium_image_link"] = df["image"].apply(lambda x: x["medium"])
```

```python
df["Original_image_link"] = df["image"].apply(lambda x: x["original"])
```

```python
df.drop("image", axis = 1, inplace = True)
```

**"summary"**

```python
df["summary"] = df["summary"].str.split('>').str[1].str.split('<').str[0]
```

**numerical features**

**"runtime"**

```python
df["runtime"] = df["runtime"].astype(float)
```

```python
##now dropping the columns whichh are not expected
df.drop(["airstamp", "_links"],axis = 1, inplace = True)
```

**Insights**

* **Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.**

```python
# Group the ratings by season and calculate the mean rating for each season
season_ratings = df.groupby("season")["rating"].mean()
```

```python
# Reset the index to convert the grouped data into a DataFrame
season_ratings = season_ratings.reset_index()
```

```python
# Sort the DataFrame by season for proper ordering
season_ratings = season_ratings.sort_values('season')
```

```python
# Plot the ratings for each season
plt.plot(season_ratings['season'], season_ratings['rating'], marker='o')

# Customize the plot
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Ratings Comparison by Season')

# Display the plot
plt.show()
```

**Get the episode name from each season with the highest and lowest rating**

```python
# Group the DataFrame by 'season'
grouped_seasons = df.groupby('season')

# Initialize empty lists to store the episode names with highest and lowest ratings
highest_rated_episodes = []
lowest_rated_episodes = []

# Iterate over each group
for season, group in grouped_seasons:
    # Find the episode with the highest rating in the current season
    highest_rated_episode = group.loc[group['rating'].idxmax(), 'name']
    highest_rated_episodes.append(highest_rated_episode)
    
    # Find the episode with the lowest rating in the current season
    lowest_rated_episode = group.loc[group['rating'].idxmin(), 'name']
    lowest_rated_episodes.append(lowest_rated_episode)

# Create a DataFrame to store the results
result_df = pd.DataFrame({'Season': grouped_seasons.groups.keys(),
                          'Highest Rated Episode': highest_rated_episodes,
                          'Lowest Rated Episode': lowest_rated_episodes})

# Print the result
print(result_df)
```

**Get the summary for the most popular ( ratings ) episode in every season**

```python
# Initialize an empty list to store the summaries
summaries = []

# Iterate over each group
for season, group in grouped_seasons:
    # Find the episode with the highest rating in the current season
    highest_rated_episode = group.loc[group['rating'].idxmax(), :]
    
    # Create a summary string with the season, episode name, and rating
    summary = f"Season {season} - Most Popular Episode: {highest_rated_episode['name']}. Rating: {highest_rated_episode['rating']}"
    
    # Append the summary to the list
    summaries.append(summary)

# Print the summaries
for summary in summaries:
    print(summary)
```
