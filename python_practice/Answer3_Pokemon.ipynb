**In the categorical columns:**
* "**type__001**" has 89 Nan values because those pokemon has only one prominent type that is mentioned in "type_001".
* the "**weaknesses**" columns have weaknesses mentioned of each pokemon. So from "weaknesses__002" if there are Nan values in those "weaknesses" columns, it is because they do not anymore weakness to mention.
* If "**prev_evolution_name**" and "**next_evolution_name**" have NaN values, it is because there are no corresponding Pokemon names in the "name" column of the dataframe to refer to for both.

So, i have considered these Nan values as separate categories in these categorical columns.

```python
# `str.replace()` is used to remove "Km" and then `pd.to_numeric()` to convert the values to numeric type. The `errors= 'coerce'` is used to convert non_numeric values to Nan
data['egg'] = pd.to_numeric(data['egg'].str.replace(' km', ''), errors='coerce')
```

```python
data["egg"].fillna(0, inplace = True)
```

```python
#handling "height" column
data["height"] = data["height"].str.split().str[0]
data["height"]= data["height"].astype(float)
```

```python
# handling "weight" column
data["weight"] = data["weight"].str.split().str[0]
data["weight"]= data["weight"].astype(float)
```

```python
##handling "spawn_time"
```

```python
data[data["spawn_time"].isna()]
```

```python
#cannot delete ros with Nan values, as may loose some potential information,
```

```python
#data["spawn_min", "spawn_sec"] = data["spawn_time"].str.split(':', expand= True)   ##gave me error that it expected 2-D array but got (151,2)
data[["spawn_min", "spawn_sec"]] = data["spawn_time"].str.split(':', expand= True)
```

```python
data["spawn_min"].fillna(0, inplace = True)
data["spawn_sec"].fillna(0, inplace = True)
```

```python
# converting the 'spawn_min' and 'spawn_sec' column to numeric type
data["spawn_min"] = data["spawn_min"].astype(int)
data["spawn_sec"] = data["spawn_sec"].astype(int)
```

```python
data.drop("spawn_time", axis = 1, inplace = True)
```

```python
data["spawn_time"] = data["spawn_min"]*60 + data["spawn_sec"]
```

```python
# encoding some of the categorical columns
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
l = [
 'type__001',
 'type__002',
 'candy',
 'weaknesses__001',
 'weaknesses__002',
 'weaknesses__003',
 'weaknesses__004',
 'weaknesses__005',
 'weaknesses__006',
 'weaknesses__007']
def encoding_categorical_columns(column_list):
    for i in column_list:
        label_encoder.fit(data[i])
        data[i] = label_encoder.transform(data[i])
```

```python
encoding_categorical_columns(l)
```

##### numerica columns

```python
## multipliers__001 and multipliers__002 are features which tells how many times powers multiply on evolution, if there is no value then may the pokemon is at their highest evolution , so safest impitaon would be 1, i.e., no change in power
data["multipliers__001"].fillna(1, inplace = True)
data["multipliers__002"].fillna(1, inplace = True)
```

```python
## "candy_count", means number of candies required for transform/evolve, as nan value could be a posiibility of requiring none as no further evolution possible
data["candy_count"].fillna(0, inplace = True)
```
